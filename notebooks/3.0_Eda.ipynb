## Question 1:

After exploring the data, we have a better understanding of the data. Now let's see what questions can be answered with this data.

**A possible question is:** Do videos with a higher popularity score on Spotify have a higher rank on YouTube?

**Answering this question will** help us to predict the rank of a video if it is published on Youtube since most of the videos nowadays appear on Spotify first.

**How we answer this question**: we will consider the `Popularity Score (Spotify)` and `Highest rank` of all music videos.

### Analyzing to get answer

Step 01: Construct dataframe for `Popularity Score (Spotify)`

```python
# Create dataframe for Popularity
popularity_df = cleaned_video_df[['Song Name', 'Popularity Score (Spotify)']]

# Sort the rows based of Popularity Score
popularity_df.sort_values('Popularity Score (Spotify)', inplace=True, ascending=False)

# Display
popularity_df.head(10)
```

```python
plt.figure(figsize=(20, 5))
sns.set(style="whitegrid")
ax = sns.barplot(data=popularity_df.head(10), x="Song Name", y= "Popularity Score (Spotify)")
```

Step 02: Construct dataframe for `Highest rank`

```python
# Create dataframe for Highest rank
rank_df = cleaned_video_df[['Song Name', 'Highest rank']]

# Sort the rows based of Highest rank
rank_df.sort_values('Highest rank', inplace=True, ascending=True)

# Display
rank_df.head(10)
```

Since all of the top 10 Highest rank are all top 1, so we will take a look at the highest rank of top 10 `Popularity Score (Spotify)`

```python
# Create dataframe for Highest rank
rank_df = cleaned_video_df[['Song Name', 'Highest rank']]

# Create a list of Song Name of Popularity Score
song_names = popularity_df['Song Name'].head(10).to_list()

# Extract song that are in top 10 of Popularity Score (Spotify)
rank_df = rank_df[rank_df['Song Name'].isin(song_names)]

# Sort the rows based of Popularity Score
rank_df.sort_values('Highest rank', inplace=True, ascending=True)

# Display rank
rank_df
```

```python
# Create a horizontal bar chart
plt.barh(rank_df['Song Name'], rank_df['Highest rank'], color='skyblue')

# Set labels and title
plt.xlabel('Rank')
plt.ylabel('Song Name')
plt.title('Song Ranks')
```

**Observations**

All I Want for Christmas Is You being placed both 1st and 2nd so there's may be 2 versions of this song. The song with the lowest score in the top 10 is We Can't Stop but its score is still more than 90. 
- However, when it comes to rank on Youtube, some songs don't make it to top 1. The two songs that reached rank 1 on Youtube are Flowers and As It Was. Despite All I Want for Christmas Is You's high score on Spotify, its music video only make to rank 3 on Youtube. And even Yellow is ranked 98 on Youtbe which is quite low. Most of the songs that ranked top 1 don't appear on top 10 Popularity Score on Spotify too.

    - Since singers usually release their music in albums and there is only 1 song is chosen as theme song, the other songs usually don't have an official music video on Youtube, they just have a lyric video so the rank of these songs on Youtube are usually not high.
    - Since Youtube has a lot of kind of videos, if the singer is not famous enough, their music videos usually don't have a high rank.

## Question 2:

**A possible question is:** What kind of music genre is likely to have a high views on Youtube or have a high Popularity Score on Spotify.

**Answering this question will** help us to know which music genre that people usually listen to. This can help the singers or composers easily to choose what music genre they should make in order to be famous.

### Analyzing to get answer

Step 01: Split each genre into a seperate row

```python
# Create new dataframe Genre
genre_df = pd.DataFrame()

# Taking out Genre column and convert its data type into list
genre_df['Genre'] = cleaned_video_df['Genre']
genre_df['Genre'] = genre_df['Genre'].str.split(',')
genre_df = genre_df.explode('Genre')
```

Step 02: Count for each genre

```python
# Count occurrences of each genre
genre_counts = genre_df['Genre'].value_counts()
genre_counts

# Create dataframe
df = pd.DataFrame(genre_counts)
df
```

## Question 3:

**A possible question is:** Who are the top artists based on the total number of views or likes and their average Popularity Score on Spotify.

**Answering this question will** help the artists to know if their music videos are good enough to be published on Spotify.

**How we answer this question**: we will consider the total `View` of each artist and count for average of `Popularity Score (Spotify)`.

### Analyzing to get answer

Step 01: Construct dataframe for total `View`

```python
# Create new datafram for total view
view_df = pd.DataFrame()

# Count sum for each artist
view_df['Total View'] = cleaned_video_df.groupby('Artist Name')['View'].sum()

# Set the index as a column - artist
view_df['Artist Name'] = view_df.index

# Drop the index
view_df.reset_index(drop=True, inplace=True)

# Sort the rows based of Total View
view_df.sort_values('Total View', inplace=True, ascending=False)

# Display
top10_view = view_df.head(10)
top10_view
```

Step 02: Construct dataframe for average `Popularity Score (Spotify)`

```python
# Create new datafram for Popularity Score
avg_popularity_df = pd.DataFrame()

# Count sum for each artist
avg_popularity_df['Avg Score'] = cleaned_video_df.groupby('Artist Name')['Popularity Score (Spotify)'].mean()

# Set the index as a column - artist
avg_popularity_df['Artist Name'] = avg_popularity_df.index

# Drop the index
avg_popularity_df.reset_index(drop=True, inplace=True)

# Sort the rows based of Total View
avg_popularity_df.sort_values('Avg Score', inplace=True, ascending=False)

# Display
avg_popularity_df.head(10)
```

Step 03: Extract top 10 views artists from `Popularity Score (Spotify)` dataframe

```python
# Create list for artist names
artist_names = view_df.head(10)['Artist Name']

# Extract artist from Popularity Score dataframe
top10_score = avg_popularity_df[avg_popularity_df['Artist Name'].isin(artist_names)]
top10_score
```

Step 04: Visualization

```python
# Merge the two datasets on 'Artist Name'
df_combined = pd.merge(top10_view, top10_score, on='Artist Name')

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(15, 6))

# Bar graph for Total Views on the primary y-axis
ax1.bar(df_combined['Artist Name'], df_combined['Total View'], color='cyan', label='Total Views')
ax1.set_xlabel('Artist Name')
ax1.set_ylabel('Total Views', color='blue')
ax1.tick_params('y', colors='blue')
ax1.legend(loc='upper left')

# Create a secondary y-axis
ax2 = ax1.twinx()

# Line graph for Average Score on the secondary y-axis
ax2.plot(df_combined['Artist Name'], df_combined['Avg Score'], marker='o', color='red', label='Average Score')
ax2.set_ylabel('Average Score', color='red')
ax2.tick_params('y', colors='red')
ax2.legend(loc='upper right')

plt.title('Total Views and Average Score by Artist')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Observations**
- Justin Bieber has the highest total views among the artists, followed by Ed Sheeran and Taylor Swift.
- On the other hand, although Taylor Swift is only placed 6th based on total views, she has the highest popularity score on Spotify.
- It can be seen that music groups don't have a high popularity score on Spotify. This is easy to understand since they focus on their music video more, their products not only focus on music but also focus on video and visual.
- Singers seem to have a better score on spotify, their score are all above 70.
- So we can conclude that to have a good score on Spotify, singers will perform better than groups and their music video on Youtube don't have to have a high views.
