## Question 3:

**A possible question is:** Who are the top artists based on the total number of views or likes and their average Popularity Score on Spotify.

**Answering this question will** help the artists to know if their music videos are good enough to be published on Spotify.

**How we answer this question**: we will consider the total `View` of each artist and count for average of `Popularity Score (Spotify)`.

### Analyzing to get answer

Step 01: Construct dataframe for total `View`

```python
# Create new datafram for total view
view_df = pd.DataFrame()

# Count sum for each artist
view_df['Total View'] = cleaned_video_df.groupby('Artist Name')['View'].sum()

# Set the index as a column - artist
view_df['Artist Name'] = view_df.index

# Drop the index
view_df.reset_index(drop=True, inplace=True)

# Sort the rows based of Total View
view_df.sort_values('Total View', inplace=True, ascending=False)

# Display
top10_view = view_df.head(10)
top10_view
```

Step 02: Construct dataframe for average `Popularity Score (Spotify)`

```python
# Create new datafram for Popularity Score
avg_popularity_df = pd.DataFrame()

# Count sum for each artist
avg_popularity_df['Avg Score'] = cleaned_video_df.groupby('Artist Name')['Popularity Score (Spotify)'].mean()

# Set the index as a column - artist
avg_popularity_df['Artist Name'] = avg_popularity_df.index

# Drop the index
avg_popularity_df.reset_index(drop=True, inplace=True)

# Sort the rows based of Total View
avg_popularity_df.sort_values('Avg Score', inplace=True, ascending=False)

# Display
avg_popularity_df.head(10)
```

Step 03: Extract top 10 views artists from `Popularity Score (Spotify)` dataframe

```python
# Create list for artist names
artist_names = view_df.head(10)['Artist Name']

# Extract artist from Popularity Score dataframe
top10_score = avg_popularity_df[avg_popularity_df['Artist Name'].isin(artist_names)]
top10_score
```

Step 04: Visualization

```python
# Merge the two datasets on 'Artist Name'
df_combined = pd.merge(top10_view, top10_score, on='Artist Name')

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(15, 6))

# Bar graph for Total Views on the primary y-axis
ax1.bar(df_combined['Artist Name'], df_combined['Total View'], color='cyan', label='Total Views')
ax1.set_xlabel('Artist Name')
ax1.set_ylabel('Total Views', color='blue')
ax1.tick_params('y', colors='blue')
ax1.legend(loc='upper left')

# Create a secondary y-axis
ax2 = ax1.twinx()

# Line graph for Average Score on the secondary y-axis
ax2.plot(df_combined['Artist Name'], df_combined['Avg Score'], marker='o', color='red', label='Average Score')
ax2.set_ylabel('Average Score', color='red')
ax2.tick_params('y', colors='red')
ax2.legend(loc='upper right')

plt.title('Total Views and Average Score by Artist')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Observations**
- Justin Bieber has the highest total views among the artists, followed by Ed Sheeran and Taylor Swift.
- On the other hand, although Taylor Swift is only placed 6th based on total views, she has the highest popularity score on Spotify.
- It can be seen that music groups don't have a high popularity score on Spotify. This is easy to understand since they focus on their music video more, their products not only focus on music but also focus on video and visual.
- Singers seem to have a better score on spotify, their score are all above 70.
- So we can conclude that to have a good score on Spotify, singers will perform better than groups and their music video on Youtube don't have to have a high views.

## Question 4:

**A possible question is:** What time frames do artists like to publish videos at, and does each time frame have any effect on the number of views on YouTube of that video?

**Answering this question will** help artists consider time frames that help increase views, or avoid time frames that can compete with many other artists.

**How we answer this question**: A day will be divided into 12 time frames. We will consider the `Publish_time`, average of `View` of each video publishing and classify them into 12 time frames. 

### Analyzing to get answer

Step 01: Construct dataframe and statistics on the number of videos posted on each time frame and the corresponding average views

```python
publish_hour_df = cleaned_video_df.loc[:, ['View', 'Publish_time']]
cleaned_video_df
publish_hour_df['Publish_time'] = pd.to_datetime(cleaned_video_df['Publish_time']).dt.hour
publish_hour_res = publish_hour_df.groupby('Publish_time', as_index=False)['View'].mean()

publish_hour_List = []

for i in range(0, 24): 
    publish_hour = publish_hour_df[publish_hour_df['Publish_time'] == i]['Publish_time'].count()
    publish_hour_List.append(publish_hour)
print(publish_hour_List)
publish_hour_res['Quantity'] = publish_hour_List

publish_hour_res
```

Step 02: Sort data in ascending/descending `Quantity` and `View` fields according to time frame.

```python
publish_hour_res.sort_values(by=['Quantity'], ascending=False).head(3)
```

```python
publish_hour_res.sort_values(by=['Quantity'], ascending=True).head(3)
```

```python
publish_hour_res.sort_values(by=['View'], ascending=False).head(3)
```

```python
publish_hour_res.sort_values(by=['View'], ascending=True).head(3)
```

Step 03: Visualization

```python
# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(15, 6))

# Bar graph for Quantity on the primary y-axis
ax1.bar(publish_hour_res['Publish_time'], publish_hour_res['Quantity'], color='green', label='Quantity')
ax1.set_xlabel('Publish_time')
ax1.set_ylabel('Quantity', color='blue')
ax1.tick_params('y', colors='green')
ax1.legend(loc='upper left')

# Create a secondary y-axis
ax2 = ax1.twinx()

# Line graph for Average View on the secondary y-axis
ax2.plot(publish_hour_res['Publish_time'], publish_hour_res['View'], marker='o', color='red', label='Average View')
ax2.set_ylabel('Average View', color='red')
ax2.tick_params('y', colors='red')
ax2.legend(loc='upper right')

plt.title('Total Views and Average View of Youtue videos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Observations**
- It shows that Videos with high views often publish videos during the hours of 4 AM, 7 AM and 4 PM; not often publish videos during the hours of 10 AM, 2 AM and 3 AM.
- At the same time, video views are higher at publishing time frames such as 3 AM, 10 PM and 11 AM; are lower at publishing time frames such as 12 AM, 6 AM and 2 AM
- From there, it shows that the time to publish videos before 8 AM is often fluctuating, more unpredictable than after 2 PM (depending on factors such as videos having international markets in many different time zones, so it is not is actually affected by local time).
- Conclusion: Video publishing time is often the initial influencing factor, and does not have a large impact on the video's final view count. However, care must also be taken to achieve the necessary initial interaction for videos that depend heavily on the domestic market.

## QUESTION 6:

**A possible question is:** How the feature `Most View Per Day` values distribute across different genres ?

**Answering this question will** give us insights into the audience'a taste on different taste of music, which could give new artists approaches on forming a particular music style in the long run for both musical values and financial income.

**How we answer this question**: We intend to categorize the genre first, then list out all the songs belong to each genre and take out their mean *most view per day*. Consequently, plot on with histogram for visualization with answer to the question.

```python
genre_df = cleaned_video_df.copy()
genre_df = genre_df.assign(genre=genre_df['Genre'].str.split(',')).explode('genre')
genre_df.drop(columns=['Genre'], inplace=True)

genre_statistic = genre_df.groupby('genre').agg({'Most view per day': 'mean'}, as_index=False).sort_values(by='Most view per day', ascending=False)
genre_statistic.reset_index(inplace=True)
#set full width
pd.set_option('display.max_colwidth', None)
genre_statistic.head(10)
```

```python
lower_quantile = genre_statistic['Most view per day'].quantile(0.25)
upper_quantile = genre_statistic['Most view per day'].quantile(0.75)
median = genre_statistic['Most view per day'].median()
```

```python
# Convert the Duration to seconds (Datetime: datetime64[ns] -> int64)
duration_popularity_df['Duration'] = duration_popularity_df['Duration'].dt.hour * 3600 + duration_popularity_df['Duration'].dt.minute * 60 + duration_popularity_df['Duration'].dt.second
```

```python

# Create a figure and axis
fig, ax = plt.subplots(figsize=(15, 10))
# Plot the scatter plot
# scale the size of the dots based on the number of views
# scale Ox by 1000 to make the graph more readable
ax.scatter(duration_popularity_df['Duration'], duration_popularity_df['Popularity Score (Spotify)'], s=duration_popularity_df['Popularity Score (Spotify)'] * 10, color='skyblue')
ax.scatter(duration_popularity_df['Duration'], duration_popularity_df['Popularity Score (Spotify)'], color='skyblue')
```

```python
# Remove outliers
duration_popularity_df = duration_popularity_df[duration_popularity_df['Duration'] < 1000]
duration_popularity_df = duration_popularity_df[duration_popularity_df['Popularity Score (Spotify)'] < 100]

# Plot the scatter plot
fig, ax = plt.subplots(figsize=(15, 10))

ax.scatter(duration_popularity_df['Duration'], duration_popularity_df['Popularity Score (Spotify)'], s=duration_popularity_df['Popularity Score (Spotify)'] * 10, color='SlateBlue', alpha=0.5)

# Set the labels and title
ax.set_xlabel('Duration (seconds)', color='SlateBlue', fontsize=15, labelpad=20, fontfamily= 'monospace')
ax.set_ylabel('Popularity Score (Spotify)', color='SlateBlue', fontsize=15, labelpad=20, fontfamily= 'monospace')
ax.set_title('Popularity Score (Spotify) by Duration', fontsize=18, fontweight='bold', fontfamily= 'monospace', color='green', loc='center', pad=20, rotation=0, wrap=True)

# draw the regression line
z = np.polyfit(duration_popularity_df['Duration'], duration_popularity_df['Popularity Score (Spotify)'], 1)
p = np.poly1d(z)
plt.plot(duration_popularity_df['Duration'], p(duration_popularity_df['Duration']), color='red', linewidth=3)

# draw the correlation coefficient
corr = duration_popularity_df['Duration'].corr(duration_popularity_df['Popularity Score (Spotify)'])
plt.text(650, 90, 'Correlation coefficient: ' + str(round(corr, 2)), fontsize=14, fontweight='bold', fontfamily= 'monospace', color='red')

# circle the dense area
plt.axvspan(0, 500, facecolor='g', alpha=0.1)
plt.axvspan(500, 1000, facecolor='y', alpha=0.1)

# set the x-axis scale
plt.xlim(0, 1000)
# set the y-axis scale
plt.ylim(0, 100)


rect = plt.Rectangle((150, 45), 250, 48, fill=False, color='orange', linewidth=2, linestyle='--')
ax.add_patch(rect)


```
