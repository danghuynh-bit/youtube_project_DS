üëÄ `Most-view-date`, `Publish_time` and `Release Date (Spotify)` are problaly date-time data type.\
üëÄ We convert these 2 columns into ***date-time***. 

```python
Most_view_date = []
Publish_time = []
Release_date = []

date_format = ['%Y', '%Y-%m', '%Y-%m-%d']

for date, time, release_date in zip(raw_df['Most-view-date'], raw_df['Publish_time'], raw_df['Release Date (Spotify)']):
    date = np.nan if pd.isna(date) else datetime.strptime(date, '%Y/%m/%d')
    time = np.nan if pd.isna(time) else datetime.strptime(time, '%Y-%m-%d %H:%M:%S')
    if (pd.isna(release_date)):
        release_date = np.nan
    elif ('-' not in (release_date)):
        release_date = datetime.strptime(release_date, '%Y')
    elif (len(release_date) == 7):
        release_date = datetime.strptime(release_date, '%Y-%m')
    else:
        release_date = datetime.strptime(release_date, '%Y-%m-%d')
    
    Most_view_date.append(date)
    Publish_time.append(time)
    Release_date.append(release_date)
    
raw_df['Most-view-date'] = Most_view_date
raw_df['Publish_time'] = Publish_time
raw_df['Release Date (Spotify)'] = Release_date
```

‚ùÑÔ∏è  For columns with numeric data types, we will calculate:   
- `Percentage` (from 0 to 100) of missing values
- The `min`
- The `lower quartile` (ph√¢n v·ªã 25)
- The `median` (ph√¢n v·ªã 50)
- The `upper quartile` (ph√¢n v·ªã 75)
- The `max`

‚ùÑÔ∏è  First we need to select columns with numeric data type.

```python
num_col_info_df = raw_df.select_dtypes(include='number')
```

```python
def missing_ratio(s):
    return (s.isna().mean() * 100).round(1)

def median(df):
    return (df.quantile(0.5)).round(1)

def lower_quartile(df):
    return (df.quantile(0.25)).round(1)

def upper_quartile(df):
    return (df.quantile(0.75)).round(1)
```

```python
num_col_info_df = num_col_info_df.agg([missing_ratio, "min", lower_quartile, median, upper_quartile, "max"])
num_col_info_df
```

üìà `COMMENT:`\
‚ùÑÔ∏è  After identifying the basic statistical numbers that describe your data, we further need to determine the features that have a large number of missing values. Such features are not useful for the analysis stage and must be removed from the dataset.   
‚ùÑÔ∏è  Usually, if the percentage of missing values is greater than 75%, the column is dropped from the dataframe and an updated dataframe is returned.

```python
def drop_missing_features(df: pd.DataFrame, missing_lst: dict = dict(num_col_info_df.iloc[0]), threshold: float = 75.0) -> pd.DataFrame:
    if (df is None) or (missing_lst is None) or (threshold is None):
        print(f"[LOG] Invalid.")
        raise ValueError
    
    # Create a copy of the dataframe to avoid changing the original
    df_cp = df.copy()
    
    # List to maintain the columns to drop
    cols_to_trim = []
    
    for key, value in missing_lst.items():
        if float(value) > threshold:
            cols_to_trim.append(key)
            
    if len(cols_to_trim) > 0:
        # drop the columns identified using the dataframe drop() method
        df_cp = df_cp.drop(columns=cols_to_trim)
        print("[LOG] Dropped column(s): " + " ".join(cols_to_trim))
    else:
        print("[LOG] Have no column(s) to trim.")
        
    return df_cp
```

```python
raw_df = drop_missing_features(raw_df, dict(num_col_info_df.iloc[0]))
```

üìà`COMMENT:`\
‚ùÑÔ∏è  After remove features which have large missing values, our dataframe still have missing values. So that, we need to fill these missing values so that they can be used in analysis.  
‚ùÑÔ∏è  There's many ways to fill in the missing values. Here we will use the median value of each column to fill in.

```python
def filling_missing_value(df: pd.DataFrame) -> pd.DataFrame:
    df = df.replace({None: np.nan})
    
    if (df is None):
        print(f"[LOG] Invalid.")
        raise ValueError
    
    # Create a copy of the dataframe to avoid changing the original
    df_cp = df.copy()
    
    # Get the list of columns in the dataframe
    cols_list = list(df_cp.columns)
    
    # Exclude the last column - Country, Year
    cols_list.remove("Id")
    cols_list.remove("Duration")
    cols_list.remove("Channel_name")
    cols_list.remove("Publish_time")
    cols_list.remove("Country")
    cols_list.remove("Most-view-date")
    cols_list.remove("Song Name")
    cols_list.remove("Artist Name")
    cols_list.remove("Release Date (Spotify)")
    cols_list.remove("Genre")
    
    # Replace all None values with NaN, fillna only works on nans
    df_cp.fillna(value=np.nan, inplace=True)
    
    # Replace all NaN values with the mean of the column values
    for col in cols_list:
        df_cp[col].fillna((df_cp[col].median()), inplace=True)
        
    return df_cp
```

‚ú® The percentage (from 0 to 100) of each value is sorted by decreasing percentage (we do not consider missing values, the ratio is the ratio compared to the number of non-missing values): you use a dictionary to store , key is the value, value is the percentage; With the column corresponding to each type, the method is similar to above.

```python
cat_col_info_df = raw_df.select_dtypes(include='object')
```

```python
def missing_ratio(s):
    return (s.isna().mean() * 100).round(1)

def num_values(s):
    s = s.str.split(';')
    s = s.explode()
    return len(s.value_counts())

def value_ratios(s):
    s = s.str.split(';')
    s = s.explode()
    totalCount = (~s.isna()).sum()
    return ((s.value_counts()/totalCount*100).round(1)).to_dict()
```

```python
cat_col_info_df = cat_col_info_df.agg([missing_ratio, num_values, value_ratios])
cat_col_info_df
```

  9. Is the collected data reasonable ?üßÆ
</div> 

This step make sure that collected data is reasonable, before moving on to the next step. These are some simple check: 
- `checking_view_vs_Like`: In reality, viewers need to access the video and increase views to be able to choose to like it (or not). Therefore, the number of views must always be greater than (or equal to) the number of likes, but absolutely cannot be smaller.
- `checking_view_vs_MostView`: A video every day will increase its views, but there cannot be a Most view per day that is greater than the total view (because the total view itself includes Most views per day and data from the previous and subsequent days).
- `checking_most_view_date`: All dates related to video statistics (such as the date with the highest views) must occur after the video's publication date.
- `checking_valid_PopularityScore`: All value in Popularity Score (Spotify) field must in range (0, 100)

```python
def checking_view_vs_MostView(df):
    return len(df[df["View"] < df["Most view per day"]]) == 0

def checking_view_vs_Like(df):
    return len(df[df["View"] < df["Like"]]) == 0

def checking_most_view_date(df):
    return len(df[df["Publish_time"] >= df["Most-view-date"]]) == 0

def checking_valid_PopularityScore (df):
    return len(df[df["Popularity Score (Spotify)"] < 0]) == 0 and len(df[df["Popularity Score (Spotify)"] > 100]) == 0
```

`checking_view_vs_Like` show that all rows have reasonable data by this criteria.

`checking_most_view_date` show that all rows have reasonable data by this criteria.

```python
checking_most_view_date(raw_df) == True
```

`checking_valid_PopularityScore` show that all rows have reasonable data by this criteria.
